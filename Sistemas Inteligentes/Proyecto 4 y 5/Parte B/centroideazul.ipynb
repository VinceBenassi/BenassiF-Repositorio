```python
# Programa de Franco Benassi
# Proyecto #4 y 5 Sistemas Inteligentes 2020
# Ejercicio 5
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as pt
import pandas as pd
import numpy as np

d3 = np.load("/home/franco-os/Documentos/Informtica/Sistemas Inteligentes/Proyecto 4 y 5/Datos 3D/puntos_3d.npy")
tabla = pd.DataFrame(d3)
tabla.columns = ["x", "y", "z"]

figura = pt.figure()
auxiliar = figura.add_subplot(111, projection='3d')
auxiliar.scatter(tabla['x'], tabla['y'], tabla['z'], c='r', marker='o')

figura_2 = pt.figure()
auxiliar2 = figura_2.add_subplot(111, projection='3d')
auxiliar2.scatter(tabla['x'], tabla['y'], tabla['z'], c='r', marker='o')
auxiliar.scatter(tabla['x'].mean(), tabla['y'].mean(), tabla['z'].mean(), c='b', marker='o', linewidths=9)

pt.show()
```
