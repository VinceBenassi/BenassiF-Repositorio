```python
# Programa de Franco Benassi
# Proyecto #4 y 5 Sistemas Inteligentes 2020
# Ejercicio 5
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as pt

Ndata = np.load("/home/franco-os/Documentos/Informática/Sistemas Inteligentes/Proyecto 4 y 5/Datos 3D/puntos_3d.npy")
dd = pd.DataFrame(Ndata)
dd.columns = ["x","y","z"]

figura = pt.figure()
Vaux = figura.add_subplot(111, projection='3d')
Vaux.scatter(dd['x'],dd['y'],dd['z'], c='r', marker='o')

figura_2 = pt.figure()
Vaux_2 = figura_2.add_subplot(111, projection='3d')
Vaux_2.scatter(dd['x'],dd['y'],dd['z'], c='r', marker='o')
Vaux_2.scatter(dd['x'].mean(),dd['y'].mean(),dd['z'],c='b',marker='o',linewidths=9)

pt.show()
```
