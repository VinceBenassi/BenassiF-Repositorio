```python
import matplotlib.pyplot as pt
import numpy as np

class FiltroKalman(object):
    def __init__(self, cc, a, std_acc, std_meas):
        self.cc = cc
        self.a = a
        self.std_acc = std_acc
        
        
        self.A = np.matrix([[1, self.cc],[0,1]])
        self.B = np.matrix([[(self.cc**2)/2],[self.cc]])
        self.H = np.matrix([[1,0]])
        self.Q = np.matrix([[(self.cc**4)/4 ,(self.cc**3)/2],[(self.cc**3)/2,self.cc**2]])*self.std_acc**2
        self.R = std_meas**2
        self.P = np.eye(self.A.shape[1])
        self.X = np.matrix([[0],[0]])
    
    
    def prediccion(self):
        self.X = np.dot(self.A, self.X) + np.dot(self.B, self.a)
        self.P = np.dot(np.dot(self.A, self.P),self.A.T + self.Q)
        return self.X
    
    
    def actualiza(self,z):
        S = np.dot(self.H,np.dot(self.P, self.H.T)) + self.R
        K = np.dot(np.dot(self.P, self.H.T),np.linalg.inv(S))
        self.X = np.round(self.X + np.dot(K,(z - np.dot(self.H, self.X))))
        I = np.eye(self.H.shape[1])
        self.P = (I - (K * self.H)) * self.P
    
    
def inicio():
    dt = 0.1
    t = np.arange(0, 100, dt)
    pos = 0.1 * (3*t - t**2)
    a = 2
    std_acc = 0.25
    std_meas = 1.2
    kf = FiltroKalman(dt, a, std_acc, std_meas)
    
    predicciones = []
    mediciones = []
    for i in pos:
        z = kf.H * i + np.random.normal(0,50)
        mediciones.append(z.item(0))
        predicciones.append(kf.prediccion()[0])
        kf.actualiza(z.item(0))
            
    pt.figure(1) 
    pt.xlabel('Tiempo (s)', fontsize=10)
    pt.ylabel('Posicion (m)', fontsize=10)
    pt.plot(pos)
        
    pt.figure(2)
    pt.title('Filtro de Kalman', fontsize=20)
    pt.plot(t, mediciones,label='Mediciones', color='b', linewidth=0.5)
    pt.plot(t,np.array(pos),label='Real Track', color='y', linewidth=1.5)
    pt.plot(t,np.squeeze(predicciones),label='Predicciones con Filtro de Kalman', color='g', linewidth=1.5)
    pt.xlabel('Tiempo (s)', fontsize=10)
    pt.ylabel('Posicion (m)', fontsize=10)
    pt.legend()
    pt.show()

if __name__ == "__main__":
    inicio()
```
