```python
# Programa de Franco Benassi
# Proyecto #4 y 5 Sistemas Inteligentes 2020
# Ejercicio 6
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as pt
import pandas as pd

iris = load_iris()
x = iris.data[:, :2]
GrupoCluster = iris.target
Colores = {0:'yellow', 1:'purple', 2:'blue'}
Dataframe = pd.DataFrame(x, columns=["x", "y"])
Dataframe['cluster'] = GrupoCluster
Dataframe["colores"] = Dataframe.cluster.map(Colores)

kmean = KMeans(n_clusters = 3)
kmean.fit(x)

pt.scatter(x="x", y="y", c="colores", s=40, data = Dataframe)
pt.xlabel('Sepal Length')
pt.ylabel('Sepal Width')
pt.show()
```
